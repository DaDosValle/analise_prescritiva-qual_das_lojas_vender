Com o código sugerido pela ALura foram obtidos 4 dataframes diferentes, um para cada loja.

Para nossa análise um único dataframe parece mais interessante.

Com isso inserimos uma nova coluna chamada "loja" na posição 0 e integramos todos os dados através do código abaixo.

```python

loja3.insert(
    loc=0,
    column="loja",
    value="loja3"
)

loja1.insert(
    loc=0,
    column="loja",
    value="loja1"
)

loja2.insert(
    loc=0,
    column="loja",
    value="loja2"
```

```python
loja4.insert(
    loc=0,
    column="loja",
    value="loja4"
)
```

Os 4 dataframes foram concatenados em 1 e em seguido, após análise as colunas não aproveitadas ("lat", "lon", "Tipo de pagamento) serão excluídas.

```python
lojas = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)
```

```python
lojas.drop("lat", axis=1, inplace=True)
lojas.drop("lon", axis=1, inplace=True)
lojas.drop("Tipo de pagamento", axis=1, inplace=True)
```

Aqui foi realizado a conversão do tipo de dados da coluna "Data da compra" para datatime.

O objetivo com isso é permitir criar um gáfico de linha para entender o comportamento de vendas e faturamento ao longo do tempo entre lojas.

```python
lojas_grafico_linha = lojas
```

```python
lojas_grafico_linha['Data da Compra'] = pd.to_datetime(lojas_grafico_linha['Data da Compra'], format='%d/%m/%Y')
```

```python
lojas_grafico_linha['Mes_Ano'] = lojas_grafico_linha['Data da Compra'].dt.to_period('M')

vendas_mensais_lojas = lojas_grafico_linha.groupby('Mes_Ano').size()

vendas_mensais_lojas.index = vendas_mensais_lojas.index.to_timestamp()
```

```python
vendas_por_loja_mensal = lojas_linha.groupby(["Mes_Ano", "loja"]).size()
vendas_para_plotar = vendas_por_loja_mensal.unstack(fill_value=0)
vendas_para_plotar.index = vendas_para_plotar.index.to_timestamp()

vendas_para_plotar.plot(kind="line",
                        figsize=(12, 6),
                        title="Vendas das 4 lojas ao longo do tempo",
                        color=["blue", "green", "orange", "purple"],
                        xlabel="Perídodo",
                        ylabel="Número de Vendas",
                        )
```
